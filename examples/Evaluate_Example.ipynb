```python
from sklearn.datasets import load_boston
import pandas as pd

boston = load_boston()
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df['price'] = boston.target
```

```python
df.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

rf = RandomForestRegressor(n_estimators=100)
X = df.drop(columns=['price']).values
Y = df['price']
```

```python
eds.evaluate.get_cross_val_ci(rf, X, Y, loss=True, metric=mean_squared_error)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=1337, shuffle=True)
```

```python
rf.fit(x_train, y_train)
```

```python
eds.evaluate.evaluate_regressor(y_test, rf.predict(x_test))
```
