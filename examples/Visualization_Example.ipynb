```python
import os
os.chdir('..')
import eds
from sklearn.datasets import load_boston
import pandas as pd
```

```python
boston = load_boston()
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df['price'] = boston.target
```

```python
df.head()
```

```python
eds.visualize.plot_hist_distribution(df, 'CRIM')
```

```python
eds.visualize.plot_relationship(df, 'CRIM', 'price')
```

```python
# Create binary feature
df['BIN_RAD'] = df['RAD'].apply(lambda x: True if x == 1 else False)
```

```python
eds.visualize.plot_binary_feature_distribution(df, 'TAX', 'BIN_RAD')
```

```python
eds.visualize.plot_binary_feature_distribution(df, 'TAX', 'BIN_RAD', hist=True)
```

```python
df['price'] = df['price'].apply(lambda x: 1 if x > 24 else 0)
```

```python
X = df.drop(columns=['price']).values
Y = df['price'].values
```

```python
eds.visualize.plot_feature_importances(X, Y, df.columns)
```
